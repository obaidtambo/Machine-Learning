# SVM

we will apply the SVM algorithm to
the Fertility dataset. The idea,is to determine whether the
fertility level of an individual is affected by their
demographics, their environmental conditions, and
their previous medical conditions.

```python
import pandas as pd
from sklearn.svm import SVC
```

```python
data =pd.read_csv("fertility_Diagnosis.csv",header=None)
```

```python
# Split the data into X and Y, considering that
# the class label is found under the column with
# the index equal to 9. Use the following code to
# do so:


X = data.iloc[:,:9]
Y = data.iloc[:,9]
```

```python
model = SVC()
model.fit(X, Y)
```

```python
SVC(C=1.0, break_ties=False, cache_size=200,class_weight=None, coef0=0.0, decision_function_shape='ovr',degree=3,
gamma='scale', kernel='rbf',
max_iter=-1,
probability=False,
random_state=None, shrinking=True,
tol=0.001, verbose=False)
```

```python
pred = model.predict([[-0.33,0.69,0,1,1,0,0.8,0,0.88]])
print(pred)
```

the model predicts the instance's class
label as N, meaning that the fertility of the
subject has not been affected.
