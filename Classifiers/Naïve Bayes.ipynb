# Naïve Bayes

apply the Naïve Bayes algorithm to a
Fertility dataset, which aims to determine whether
the fertility level of an individual has been affected
by their demographics, their environmental
conditions, and their previous medical conditions.

```python
import pandas as pd
from sklearn.naive_bayes import GaussianNB
```

```python
data =pd.read_csv("./fertility_Diagnosis.csv",header=None)
```

```python
# Split the data into X and Y, considering that
# the class label is found under the column with
# an index equal to 9. Use the following code to
# do so:

X = data.iloc[:,:9]
Y = data.iloc[:,9]
```

```python
model = GaussianNB()
model.fit(X, Y)
```

```python
pred = model.predict([[-0.33,0.69,0,1,1,0,0.8,0,0.88]])
print(pred)
```

The predicted class for that subject is equal to N,
which means that the fertility of the subject
has not been affected.
